```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('BCASTudents_exam_performance.csv')
```

```python
#handel Missing value 
df.isnull().sum()
df[df.isnull().any(axis=1)]
```

```python
df.dropna(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df['Name'] = df['Name'].str.upper().str.strip()   
df['Name'] = df['Name'].replace({'old': 'new'})     # Fix values
```

```python
df
```

```python
df = df.drop(columns=["Result"])
```

```python
df
```

```python
subjects = ["AI", "Maths", "Web Tech", "Software Engineering (SE)"]
avg_scores = df[subjects].mean()

plt.figure(figsize=(8, 5))
plt.pie(avg_scores, labels=subjects, autopct='%1.1f%%', startangle=140)
plt.title("Average Marks per Subject")
plt.axis('equal')  
plt.show()
```

```python
df['passed'] = df['passed'].astype(bool)

total = df.groupby('Specialization').size()
passed = df[df['passed']].groupby('Specialization').size()

summary = pd.DataFrame({
    'Total Students': total,
    'Passed Students': passed
})
plt.figure(figsize=(8, 6))
bars = plt.bar(summary.index, summary['Passed Students'], color='mediumseagreen',width=0.5)
plt.xlabel("Specialization")
plt.ylabel("Number of Students Passed")
plt.yticks(rotation=10)
plt.tight_layout()

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, str(height), ha='center', fontsize=9)
  
plt.show()
```

```python
subject_columns = ['AI', 'Maths', 'Web Tech', 'Software Engineering (SE)']

average_scores_by_section = df.groupby('Section')[subject_columns].mean()

average_scores_by_section.T.plot(figsize=(10, 6), marker='o')

plt.title('Average Subject Scores by Section')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.legend(title='Section')
plt.tight_layout()
plt.show()
```

```python
subject_cols = ['AI', 'Maths', 'Web Tech', 'Software Engineering (SE)']

top_students_by_section = df.loc[df.groupby('Section')['Total Marks'].idxmax()]

plt.figure(figsize=(10, 6))

for _, row in top_students_by_section.iterrows():
    plt.plot(subject_cols, row[subject_cols], marker='o', label=f"{row['Name']} (Section {row['Section']})")

plt.title("Top Scoring Students in Each Section - Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.tight_layout()
plt.show()
```
